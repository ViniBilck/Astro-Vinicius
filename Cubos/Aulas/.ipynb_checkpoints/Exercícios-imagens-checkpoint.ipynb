## 1. Carregar imagens na memória

Primeiro passo: baixar as imagens. Lembrando, estamos trabalhando com as imagens deste frame do SDSS:

```
https://dr12.sdss.org/fields/runCamcolField?field=130&camcol=1&run=6793
```

Você pode usar o browser e baixar manualmente cada imagem, ou pode rodar o script abaixo. Depois de rodar a célula abaixo, os arquivos com as imagens devem aparecer no mesmo diretório deste notebook.

````python
import requests
from os import path

filters = ['u', 'g', 'r', 'i', 'z']
url_template = 'https://dr12.sdss.org/sas/dr12/boss/photoObj/frames/301/6793/1/frame-{}-006793-1-0130.fits.bz2'
# Mude a linha de baixo para True para forçar baixar o arquivo novamente.
force = False

for f in filters:
    url = url_template.format(f)
    # O nome do arquivo é a última parte da URL.
    fname = url.split('/')[-1]
    if not force and path.exists(fname):
        print(f'file {fname} exists, skipping download.')
        continue
    print(f'downloading {url} ...')
    with requests.get(url) as r:
        print(f'saving to {fname}.')
        with open(fname, 'wb') as w:
            w.write(r.content)
````

### Anatomia de um arquivo FITS

Os arquivos baixados são imagens no formato FITS. Esses arquivos podem incluir múltiplas imagens e tabelas (chamados *extensions*), cada uma com seu respectivo cabeçalho (*headers*) contendo mais dados e metadados.

**Nota:** repare que os arquivos têm uma *extensão* `.bz2`. Isso significa que os arquivos estão compactados com o programa `bzip2`. Isso não faz diferença pra nós no momento, pois vamos ler os arquivos usando o pacote `astropy`, que descompacta automaticamente.

Neste nosso exercício vamos trabalhar apenas com a primeira extensão, que contém uma imagem com o fluxo do campo observado. Para carregar as imagens, usamos o módulo `astropy.io.fits`, que contém (entre outras coisas) as funções `getdata()` e `getheader()`.

#### Header

No cabeçalho destas imagens existem informações relacionadas ao sistema de coordenadas, que permitem traduzir a posição de cada pixel a uma posição física no céu. Existem ainda dados sobre o observatório, data de aquisição, tempo de exposição, e mais um mundaréu de coisas relacionadas à *pipeline* de tratamento de dados por onde esta imagem passou.

Um header funciona como um dicionário, com palavras-chave e valores (mais comentários). Não vamos precisar de nenhuma informação do header aqui, mas convém se acostumar com o que tem dentro. Vamos carregar o header da imagem na banda `r`.

````python
from astropy.io import fits

h = fits.getheader('frame-r-006793-1-0130.fits.bz2')
data_obs = h['BUNIT']
print(f'{data_obs}')
````

### Escalas de um gráfico de imagem
Deveria haver uma galáxia bonita ali, como aparece no site. O que deu errado? Para entender isso, precisamos entender como essa figura é gerada.

As imagens que temos aqui são monocromáticas. A função `imshow()` precisa arrumar um jeito de escolher qual cor desenhar cada pixel. Ela faz isso usando uma coisa chamada *mapa de cores*, ou *colormap*. O colormap aparece na escala de cores à direita da figura (*colorbar*). Tecnicamente, precisamos de uma forma de mapear os valores dos pixels em valores de 0 até 1, que correspondem às cores do colormap. Por padrão, o `imshow()` usa um mapeamento linear, como pode ser visto no gráfico abaixo:

````python
N = 100
i = 50
vmin, vmax = 2, 9
x = np.linspace(vmin, vmax, N)
y = np.linspace(0, 1, N)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot([0, vmax, vmax], [1, 1, 0], 'k--')
ax.plot([0, x[i]], [y[i], y[i]], 'k--')
ax.scatter(x[i], y[i], c='r', s=100)
ax.plot([0, x[i], x[i]], [y[i], y[i], 0], 'k--')
ax.set_ylim(0, 1.1)
ax.set_yticks([0, y[i], 1])
ax.set_yticklabels([r'$0$', r'$y$', r'$1$'])
ax.set_ylabel('valores mapeados')
ax.set_xlim(0, vmax + 1)
ax.set_xticks([vmin, x[i], vmax])
ax.set_xticklabels([r'$v_\mathrm{min}$', r'$x$', r'$v_\mathrm{max}$'])
ax.set_xlabel('valores dos pixels')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
````

Quer dizer, supondo que temos valores dos pixels entre $v_\text{min}$ e $v_\text{max}$, podemos mapear um pixel com valor $x$ linearmente na faixa entre 0 e 1 resolvendo para $y$ que está sobre a reta. Pra quem ainda não fez álgebra linear,

$$
y = \frac{x - v_\text{min}}{v_\text{max} - v_\text{min}}.
$$

Note que é necessário saber $v_\text{min}$ e $v_\text{max}$. Novamente, por padrão a função `imshow()` escolhe os valores automaticamente. Esses valores automáticos são os mínimos e os máximos presentes na imagem, o que é satisfatório em boa parte dos casos. Não parece ser o nosso caso, então precisamos escolher os valores mínimo e máximo, usando os argumentos `vmin` e `vmax`. Primeiro, vamos repetir o gráfico com máximo (`np.max()`) e mínimo (`np.min()`) absolutos, para saber a sintaxe correta. Confira se a imagem é igual à anterior.

````python
fig, ax = plt.subplots()
im = ax.imshow(im_r, vmin=np.min(im_r), vmax=np.max(im_r))
plt.colorbar(im, label='nanomaggies')
plt.show()
````

A tentação agora é ficar na base da tentativa e erro, buscando valores para colocar em `vmax` e `vmin`. Vamos pelo menos automatizar isso usando ferramentas chiques do jupyter. Abaixo fazemos um gráfico interativo, onde podemos escolher os valores de `vmax` e `vmin` com um controle deslizante.

````python
import ipywidgets as widgets
from IPython.display import display

def plota_imagem(vmin, vmax):
    fig, ax = plt.subplots()
    im = ax.imshow(im_r, vmin=vmin, vmax=vmax)
    plt.colorbar(im, label='nanomaggies')
    plt.show()
        
slmin = widgets.FloatSlider(value=np.min(im_r), min=np.min(im_r), max=np.max(im_r),
                           description='vmin')
slmax = widgets.FloatSlider(value=np.max(im_r), min=np.min(im_r), max=np.max(im_r),
                           description='vmax')
grafico = widgets.interactive_output(plota_imagem, {'vmin': slmin, 'vmax':slmax})

display(grafico)
display(slmin)
display(slmax)
````

Brinque com os valores, e descubra uma faixa melhor. Você deve encontrar facilmente esses valores, e vai ver que eles são bem diferentes do e mínimo máximo absoluto usado por padrão. Precisamos de uma forma melhor de escolher `vmax` e `vmin`.

### Estatísticas de pixels

Em vez de sair nessa busca cega, vamos entender os nossos dados. Temos uma coleção de valores de pixels, e queremos escolher `vmin` e `vmax` tais que a escala de cores seja bem utilizada. Uma forma de fazer isso é usar valores mais frequentes. Podemos buscar valores próximos da média aritmética do fluxo dos pixels. As funções `np.mean()` e `np.std()` são úteis aqui, calculam respectivamente a média aritmética e o desvio quadrático médio da média.

````python
print(f'média: {np.mean(im_r)}')
print(f'desvio padrão: {np.std(im_r)}')
````

Veja que os valores são bem menores do que o máximo que usamos, acima de 200. Na verdade, a grande maioria dos pontos tem valores próximos de zero, e são menores do que 1.

#### Entendendo melhor a distribuição de pixels

A média e o desvio padrão, usados acima, servem para caracterizar resumidamente distribuições.
Vamos analisar com mais cuidado a distribuição de valores dos pixels, isto é, quantos pixels de cada valor existem. Como os valores são números reais, o correto é tratarmos a distribuição como uma densidade, ou seja, saber quantos pixels temos entre um valor $x$ e um valor $x + \Delta x$. A forma que vamos usar para conhecer essa distribuição é através de um histograma. Num histograma, definimos intervalos (ou caixas, ou *bins*) que não se sobrepõem, e contamos quantos pixels têm valor dentro de cada caixa. Cada pixel cai dentro de somente uma caixa. Com um número grande de pixels, e bin bem estreitos, o histograma se aproxima bem da densidade de distribuição de valores de pixels. 

Ainda não conhecemos a distribuição, sabemos apenas qual o seu valor máximo, o mínimo, a média e o desvio padrão. Vamos Então desenhar o histograma usando a função `ax.hist()`. Precisamos definir os bins, que podem ser feitos explicitamente, ou simplesmente dizemos quantos queremos, e em qual faixa de valores. Faremos do segundo modo. também vamos usar

````python
fig, ax = plt.subplots()

# Precisamos de um array unidimensional para calcular o histograma.
im_r_flat = im_r.flatten()

ax.hist(im_r_flat, bins=50, range=[np.min(im_r), np.max(im_r)])
ax.set_xlabel('nanomaggies')
plt.show()
````

Não, isso não foi um erro. Como vimos antes, a maioria dos pixels tem valor menor do que 1, mas deve haver alguns poucos com valor muito alto. Vamos então usar novamente o gráfico interativo, para tentar escolher uma faixa mais representativa.

````python
import ipywidgets as widgets
from IPython.display import display

def plota_hist(vmin, vmax):
    fig, ax = plt.subplots()
    ax.hist(im_r_flat, bins=50, range=[vmin, vmax])
    plt.show()
        
slmin = widgets.FloatSlider(value=np.min(im_r), min=np.min(im_r), max=np.max(im_r),
                           description='vmin')
slmax = widgets.FloatSlider(value=np.max(im_r), min=np.min(im_r), max=np.max(im_r),
                           description='vmax')
grafico = widgets.interactive_output(plota_hist, {'vmin': slmin, 'vmax':slmax})

display(grafico)
display(slmin)
display(slmax)
````
