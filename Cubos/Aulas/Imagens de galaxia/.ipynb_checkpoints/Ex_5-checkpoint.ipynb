```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-bright')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.linewidth'] = 1.5
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['image.origin'] = 'lower'
plt.rcParams['image.cmap'] = 'viridis'

```

```python
# Repetindo a lista de filtros para facilitar a visualização.
filters = ['u', 'g', 'r', 'i', 'z']
filters_wl = [3551.0, 4686.0, 6166.0, 7480.0, 8932.0]

data_fits = np.array([fits.getdata(f'frame-{_}-006793-1-0130.fits.bz2') for _ in filters]) 
```

```python
def fun(x0, y0, data):
    d = 10
    x1 = x0 - d
    x2 = x0 + d
    y1 = y0 - d
    y2 = y0 + d
    
    flux_sum=[] 
    for i in range(len(data)):
        im_c = data[i][y1:y2, x1:x2]
        sum_imcflat = np.sum(im_c)
        flux_sum.append(sum_imcflat)
    return flux_sum




```

```python
#Plot com somente do centro 
centro = fun(1515, 977, data_fits)
fig, ax = plt.subplots()
plt.plot(filters_wl,centro, 'o-', label='Centro', color='Orange')
ax.set_xlabel('Wavelength (A)')
ax.set_ylabel('Nanomaggie')

plt.legend()
plt.show()
```

```python
#Plot com a normalização 
centro = fun(1515, 977,data_fits)
deslocado_150 = fun(1615, 977,data_fits)
fig, ax = plt.subplots()
plt.plot(filters_wl,centro/centro[2], 'o-', label='Centro', color='Orange')
plt.plot(filters_wl,deslocado_150/deslocado_150[2],'o-', label='Braço', color='#338AFF')
ax.set_xlabel('Wavelength (A)')
ax.set_ylabel('Adimensional')
plt.legend()
plt.show()
```
